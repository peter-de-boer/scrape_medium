Merge the dataframes:

```python
frames = [different_kind_of_adjective_things,different_kind_of_things,different_kind_of_things_words]
nthings = pd.concat(frames)
nthings
```

```python
def showStats(df):
    sizes = df.groupby('searchString', sort=False).size()
    for i, v in sizes.items():
        print('index: ', i, 'value: ', v)
```

```python
thing='thing'
matchString = '\d ' + thing
things = nthings.loc[nthings['searchString'].str.contains(matchString)]
```

```python
showStats(things)
```

```python
def showAllTitles(df, limit=None, showTotalClapCount=False):
    count=0
    if showTotalClapCount:
            print('{:>20}   {}'.format('totalClapCount', 'title'))
    for index, row in df.iterrows():
        if showTotalClapCount:
            print('{:>20}   {}'.format(row['totalClapCount'], row['title']))
        else:
            print(row['title'])
        count += 1
        if limit and count>=limit:
            break
        

def showTitlesSingleString(df, searchString, limit = None):
    # print all (if limit==None) or at most limit titles for the given searchString
    subset = df[df['searchString']==searchString]
    showAllTitles(subset, limit)

        
def showTitles(df, searchString, limit=None):
    if isinstance(searchString, list):
        for s in searchString:
            showTitlesSingleString(df, s, limit)
    else:
        showTitlesSingleString(df, searchString, limit)
```
